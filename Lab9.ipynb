```python
#W9A1
a = list(map(int, input().split()))
b = list(map(int, input().split()))
i = j = 0
res = []
while i < len(a) and j < len(b):
  if a[i] < b[j]:
    res.append(a[i])
    i += 1
  else:
    res.append(b[j])
    j += 1
res.extend(a[i:])
res.extend(b[j:])
print(*res)
```

```python
#W9A7
def canPlaceFlowers(flowerbed, k):
  v = 0
  if len(flowerbed) >= 3:
    for i in range(1, len(flowerbed) -1):
      if flowerbed[i - 1] == 0 and flowerbed[i] == 0 and flowerbed[i + 1] == 0:
        v += 1
  else:
    if 1 not in flowerbed:
      v += 2
  return v >= k
flowerbed = list(map(int, input().split()))
k = int(input())
print(canPlaceFlowers(flowerbed, k))
```

```python
#W9A8
def is_path_crossing(moves):
  toa_do_hien_tai = [0,0]
  toa_do_da_di = [[0,0]]
  cross = False
  for i in moves:
    if i == 'R':
      toa_do_hien_tai[0] += 1
    elif i == 'L':
      toa_do_hien_tai[0] -= 1
    elif i == 'U':
      toa_do_hien_tai[1] += 1
    elif i == 'D':
      toa_do_hien_tai[1] -= 1
    toa_do_da_di.append(toa_do_hien_tai.copy())
  for k in range(len(toa_do_da_di) - 1):
    for j in range((k+1),len(toa_do_da_di)):
      if toa_do_da_di[k] == toa_do_da_di[j]:
        cross = True
  return cross
a = input().strip()
print(is_path_crossing(a))
```

```python
#W9A10
import math
houses = list(map(int, input().split()))
heaters = list(map(int, input().split()))
def findRadius(houses, heaters):
  i = 0
  ans = 0
  for house in houses:
    while i + 1 < len(heaters) and heaters[i + 1] <= house:
      i += 1
    left_dist = abs(house - heaters[i])
    right_dist = float('inf')
    if i + 1 < len(heaters):
      right_dist = abs(house - heaters[i + 1])
    r = min(left_dist, right_dist)
    ans = max(ans, r)
  return ans


print(findRadius(houses, heaters))
```
