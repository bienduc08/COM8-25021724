```python
#W6A1
def rieng(s):
  rieng=[]
  for i in range (0,len(s)):
      if s[i] not in rieng:
        rieng.append(s[i])
  return rieng
a=list(map(int,input().split()))
print(rieng(a))
```

```python
#W6A2
def tong(s):
  tong=[]
  for i in range (0,len(s)):
    tong.append(sum(s[0:i+1]))
  return tong
a=list(map(int,input().split()))
print(tong(a))
```

```python
#W6A9
import ast
def tong(d1 : dict, d2: dict)-> dict[str,int]:
    out = dict(d1)
    for k,v in d2.items():
      if k in out:
        out[k] = out.get(k) + v
      elif k not in out:
        out[k]=v
    return out
d1=ast.literal_eval(input())
d2=ast.literal_eval(input())
print(tong(d1,d2))
```
