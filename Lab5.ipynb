```python
#W5A3
import math
def prime(a):
  if a<2 :return False
  for i in range (2,int(math.sqrt(a))+1):
    if a%i==0 :
      return False
  return True
a =int (input())
print(prime(a))
```

```python
#W5A5
from typing import List

def first_index(lst: List[int], k: int) -> int:
    # trả về vị trí 1-based, nếu không có trả về -1
    for i, val in enumerate(lst, start=1):
        if val == k:
            return i
    return -1
print(first_index(lst,k))
```

```python
#W5A6
def giaithua(a):
  tich=1
  if a==0:
    return 1
  elif a>0:
    for i in range (1,a+1):
      tich=tich*i
    return tich
m =int(input())
print(giaithua(m))


```

```python
#W5A7
def nhap():
    return float(input()), input().strip(), float(input())
def calculator(num1: float, operat: str, num2: float) -> float:
    if operat == '+':
        res = num1 + num2
    elif operat == '-':
        res = num1 - num2
    elif operat == '*':
        res = num1 * num2
    elif operat == '/':
        if num2 == 0:
            raise ZeroDivisionError("Division by zero")
        res = num1 / num2
    else:
        raise ValueError("Unsupported operator")
    # result rounded to 2 decimal places
    return f"{res:.2f}"
a,b,c=nhap()
print(calculator(a,b,c))
```

```python
#W5A8
def hamming_distance(x: int, y: int) -> int:
    return bin(x ^ y).count('1')

a,b=map(int,input().split())
print(hamming_distance(a,b))
```

```python
#W5A9
def sum_digits(n: int) -> int:
    n = abs(n)
    s = 0
    while n:
        s += n % 10
        n //= 10
    return s
a=int(input())
print(sum_digits(a))
```

```python
#W5A10
def is_isomorphic(a: str, b: str) -> bool:
    if len(a) != len(b):
        return False
    map_ab = {}
    map_ba = {}
    for ca, cb in zip(a, b):
        if ca in map_ab:
            if map_ab[ca] != cb:
                return False
        else:
            map_ab[ca] = cb
        if cb in map_ba:
            if map_ba[cb] != ca:
                return False
        else:
            map_ba[cb] = ca
    return True
a,b=map(str,input().split())
print(is_isomorphic(a,b))
```
