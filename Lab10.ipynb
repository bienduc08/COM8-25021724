```python
#W10A3
def is_happy(n):
    seen = set()
    while n != 1:
        if n in seen:
            return False
        seen.add(n)
        n = sum(int(d)**2 for d in str(n))
    return True

try:
    a = int(input())
    print("YES" if is_happy(a) else "NO")
except ValueError:
    print("Nhap sai kieu du lieu")
```
