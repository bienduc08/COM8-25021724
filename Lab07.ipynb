```python
#W7A1
def binary_search(arr,target):
  p=0
  r=0
  for i in range (len(arr)):
    if arr[i]==target:
      p=i
      r=1
  if r==1:
    return p
  else:
    return -1
arr=list(map(int,input().split()))
target=int(input())
print(binary_search(arr,target))
```

```python
#W7A3
def bubble_sort(arr):
    a = arr[:]
    n = len(a)
    for i in range(n):
        for j in range(0, n - i - 1):
            if a[j] > a[j + 1]:
                a[j], a[j + 1] = a[j + 1], a[j]
    return a

def selection_sort(arr):
    a = arr[:]
    n = len(a)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if a[j] < a[min_idx]:
                min_idx = j
        a[i], a[min_idx] = a[min_idx], a[i]
    return a
arr=list(map(int,input().split()))
print(bubble_sort(arr))
print(selection_sort(arr))
```

```python
#W7A4
def most_frequent(arr):
    freq = {}
    for x in arr:
        freq[x] = freq.get(x, 0) + 1

    max_count = max(freq.values())
    for x in arr:
        if freq[x] == max_count:
            return x, max_count
arr=list(map(int,input().split()))
print(most_frequent(arr))
```

```python
#W7A5
def count_pairs(arr, X):
    seen = {}
    count = 0
    for x in arr:
        count += seen.get(X - x, 0)
        seen[x] = seen.get(x, 0) + 1
    return count
arr=list(map(int,input().split()))
X=int(input())
print(count_pairs(arr,X))
```

```python
#W7A8
import ast

def min_interval(intervals, queries):
    res = []
    for q in queries:
        best = float('inf')
        for l, r in intervals:
            if l <= q <= r:
                best = min(best, r - l + 1)
        res.append(best if best != float('inf') else -1)
    return res
intervals=ast.literal_eval(input())
queries=ast.literal_eval(input())
print(min_interval(intervals,queries))
```

```python
import bisect

def min_interval(intervals, queries):
    n = len(queries)
    res = [-1] * n

    q_sorted = sorted((q, i) for i, q in enumerate(queries))
    qs = [q for q, _ in q_sorted]

    best = [float('inf')] * n

    for l, r in intervals:
        length = r - l + 1
        left = bisect.bisect_left(qs, l)
        right = bisect.bisect_right(qs, r)

        for i in range(left, right):
            best[q_sorted[i][1]] = min(best[q_sorted[i][1]], length)

    for i in range(n):
        res[i] = best[i] if best[i] != float('inf') else -1

    return res
intervals=ast.literal_eval(input())
queries=ast.literal_eval(input())
print(min_interval(intervals,queries))
```

```python
#W7A10
import bisect
import ast

def max_subarray_mod(nums, m):
    prefix = 0
    prefix_set = []
    result = 0

    for x in nums:
        prefix = (prefix + int(x)) % m
        result = max(result, prefix)
        i = bisect.bisect_right(prefix_set, prefix)
        if i < len(prefix_set):
            result = max(result, (prefix - prefix_set[i] + m) % m)
        bisect.insort(prefix_set, prefix)

    return result
nums=ast.literal_eval(input())
m=int(input())
print(max_subarray_mod(nums,m))
```

```python
#W7A12
def dictionary_order(words):
    indexed = list(enumerate(words))
    indexed.sort(key=lambda x: x[1].lower())
    return " ".join(chr(65 + i) for i, _ in indexed)
words=input()
print(dictionary_order(words))
```

```python
#W7A13
def harmonious(arr):
    from collections import Counter
    cnt = Counter(arr)
    ans = 0
    for x in cnt:
        if x + 1 in cnt:
            ans = max(ans, cnt[x] + cnt[x + 1])
    return ans
arr=list(map(int,input().split()))
print(harmonious(arr))
```

```python
#W7A14
def find_seven(arr):
    pos = [i for i in range(len(arr)) if arr[i] == 7]
    if not pos:
        print("Not found")
    else:
        print(" ".join(map(str, pos[::-1])))
a=int(input())
for _ in range(a):
    arr=list(map(int,input().split()))
    find_seven(arr)
```
