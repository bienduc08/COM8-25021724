```python
#W8A1
import math

class Cylinder:
    def __init__(self, r, h):
        self.r = r
        self.h = h

    def surface_area(self):
        return 2 * 3.14 * self.r * (self.r + self.h)

    def volume(self):
        return 3.14 * self.r * self.r * self.h


r, h = map(float, input().split())
c = Cylinder(r, h)

print(f"{c.surface_area():.2f} {c.volume():.2f}")

```

```python
#W8A2
class Date:
    def __init__(self, d, m, y):
        self.d = d
        self.m = m
        self.y = y

    def is_leap(self):
        return self.y % 400 == 0 or (self.y % 4 == 0 and self.y % 100 != 0)

    def is_valid(self):
        days = [31, 29 if self.is_leap() else 28, 31, 30, 31, 30,
                31, 31, 30, 31, 30, 31]
        return 1 <= self.m <= 12 and 1 <= self.d <= days[self.m - 1]

    def next_day(self):
        days = [31, 29 if self.is_leap() else 28, 31, 30, 31, 30,
                31, 31, 30, 31, 30, 31]
        self.d += 1
        if self.d > days[self.m - 1]:
            self.d = 1
            self.m += 1
            if self.m > 12:
                self.m = 1
                self.y += 1


s = input().strip()
d, m, y = map(int, s.split('/'))
date = Date(d, m, y)

if not date.is_valid():
    print("INVALID")
else:
    date.next_day()
    print(f"{date.d:02d}/{date.m:02d}/{date.y}")
```

```python
#W8A3
import math

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def is_origin(self):
        return self.x == 0 and self.y == 0

    def on_x_axis(self):
        return self.y == 0 and self.x != 0

    def on_y_axis(self):
        return self.x == 0 and self.y != 0

    def distance_to_origin(self):
        return math.sqrt(self.x**2 + self.y**2)


x, y = map(float, input().split())
p = Point(x, y)

if p.is_origin():
    print("M la goc toa do")
elif p.on_y_axis():
    print("M nam tren truc tung")
elif p.on_x_axis():
    print("M nam tren truc hoanh")
else:
    print("M khong nam tren truc toa do")

print(f"{p.distance_to_origin():.2f}")
```

```python
#W8A4
class Calculator:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    def set_numbers(self, a, b):
        self.a = a
        self.b = b

    def add(self): return self.a + self.b
    def subtract(self): return self.a - self.b
    def product(self): return self.a * self.b
    def divide(self): return self.a / self.b if self.b != 0 else "Error"
    def power(self): return self.a ** self.b
    def mod(self): return self.a % self.b


a, b = map(float, input().split())
calc = Calculator(a, b)

while True:
    op = input().strip()
    if op == "1":
        print(calc.add())
    elif op == "2":
        print(calc.subtract())
    elif op == "3":
        print(calc.product())
    elif op == "4":
        print(calc.divide())
    elif op == "5":
        print(calc.power())
    elif op == "6":
        print(calc.mod())
    elif op == "7":
        a, b = map(float, input().split())
        calc.set_numbers(a, b)

    exit_q = input("Ban co muon thoat? ")
    if exit_q.lower() == "yes":
        break
```

```python
#W8A5
class ShoppingCart:
    def __init__(self):
        self.items = []

    def add(self, name, price):
        self.items.append({"name": name, "price": price})

    def remove(self, name):
        self.items = [i for i in self.items if i["name"] != name]

    def is_empty(self):
        return len(self.items) == 0

    def total(self):
        return sum(i["price"] for i in self.items)

    def show(self):
        for i in self.items:
            print(i["name"], i["price"])

    def clear(self):
        self.items.clear()


cart = ShoppingCart()

while True:
    cmd = input().split()
    if cmd[0] == "2":
        cart.add(cmd[1], float(cmd[2]))
    elif cmd[0] == "3":
        cart.remove(cmd[1])
    elif cmd[0] == "4":
        print(cart.is_empty())
    elif cmd[0] == "5":
        print(cart.total())
    elif cmd[0] == "6":
        cart.show()
    elif cmd[0] == "7":
        cart.clear()

    if input().lower() == "yes":
        break
```

```python
#W8A6
class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h

    def area(self):
        return self.w * self.h

    def perimeter(self):
        return 2 * (self.w + self.h)

    def scale(self, k):
        self.w *= k
        self.h *= k


w, h, k = map(float, input().split())
r = Rectangle(w, h)
r.scale(k)

print(f"{r.area():.2f} {r.perimeter():.2f}")
```

```python
#w8A7
from math import gcd

class Fraction:
    def __init__(self, num, den):
        if den < 0:
            num, den = -num, -den
        g = gcd(abs(num), abs(den))
        self.num = num // g
        self.den = den // g

    def add(self, other):
        return Fraction(self.num*other.den + other.num*self.den, self.den*other.den)

    def sub(self, other):
        return Fraction(self.num*other.den - other.num*self.den, self.den*other.den)

    def mul(self, other):
        return Fraction(self.num*other.num, self.den*other.den)

    def div(self, other):
        return Fraction(self.num*other.den, self.den*other.num)

    def __str__(self):
        return f"{self.num}/{self.den}"


a, b, op, c, d = input().split()
f1 = Fraction(int(a), int(b))
f2 = Fraction(int(c), int(d))

res = {"+" : f1.add, "-" : f1.sub, "*" : f1.mul, "/" : f1.div}[op](f2)
print(res)
```

```python
#W8A8
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def deposit(self, x):
        self.balance += x

    def withdraw(self, x):
        if x <= self.balance:
            self.balance -= x


owner, balance = input().split()
acc = BankAccount(owner, float(balance))
n = int(input())

for _ in range(n):
    cmd, x = input().split()
    x = float(x)
    if cmd == "DEPOSIT":
        acc.deposit(x)
    else:
        acc.withdraw(x)

print(f"{acc.balance:.2f}")
```

```python
#W8A9
class Student:
    def __init__(self, name):
        self.name = name
        self.scores = []

    def add_score(self, subject, score):
        self.scores.append(score)

    def avg(self):
        return sum(self.scores) / len(self.scores)

    def rank(self):
        a = self.avg()
        if a >= 8: return "Excellent"
        if a >= 6.5: return "Good"
        if a >= 5: return "Average"
        return "Poor"


name = input()
n = int(input())
s = Student(name)

for _ in range(n):
    _, score = input().split()
    s.add_score(_, float(score))

print(f"{s.name} {s.avg():.2f} {s.rank()}")
```

```python
#W8A10
class Book:
    def __init__(self, t, a, y):
        self.title = t
        self.author = a
        self.year = y


class Library:
    def __init__(self):
        self.books = []

    def add(self, b):
        self.books.append(b)

    def count_by_author(self, name):
        return sum(1 for b in self.books if b.author == name)

    def find_by_year(self, y):
        return sum(1 for b in self.books if b.year == y)


lib = Library()
n = int(input())

for _ in range(n):
    cmd = input()
    if cmd.startswith("ADD"):
        t, a, y = cmd[4:].split(";")
        lib.add(Book(t, a, int(y)))
    elif cmd.startswith("COUNTYEAR"):
        print(lib.find_by_year(int(cmd.split()[1])))
    else:
        print(lib.count_by_author(cmd.split()[1]))
```

```python
#W8A11
import math

def vec(a, b):
    return (b[0]-a[0], b[1]-a[1], b[2]-a[2])

def cross(u, v):
    return (
        u[1]*v[2] - u[2]*v[1],
        u[2]*v[0] - u[0]*v[2],
        u[0]*v[1] - u[1]*v[0]
    )

def dot(u, v):
    return u[0]*v[0] + u[1]*v[1] + u[2]*v[2]

def norm(u):
    return math.sqrt(dot(u, u))


A = tuple(map(float, input().split()))
B = tuple(map(float, input().split()))
C = tuple(map(float, input().split()))
D = tuple(map(float, input().split()))

AB = vec(A, B)
AC = vec(A, C)
BC = vec(B, C)
BD = vec(B, D)

n1 = cross(AB, AC)
n2 = cross(BC, BD)

cos_phi = abs(dot(n1, n2)) / (norm(n1) * norm(n2))
phi = math.degrees(math.acos(cos_phi))

print(f"{phi:.2f}")
```

```python
#W8A12
class ComplexNumber:
    def __init__(self, r, i):
        self.r = r
        self.i = i

    def multiply(self, other):
        real = self.r * other.r - self.i * other.i
        imag = self.r * other.i + self.i * other.r
        return ComplexNumber(real, imag)


a, b = map(float, input().split())
c, d = map(float, input().split())

z1 = ComplexNumber(a, b)
z2 = ComplexNumber(c, d)

res = z1.multiply(z2)
print(res.r, res.i)
```

```python
#W8A13
import math

class Triangle:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    def getArea(self):
        if self.a + self.b <= self.c or self.a + self.c <= self.b or self.b + self.c <= self.a:
            return -1
        p = (self.a + self.b + self.c) / 2
        return math.sqrt(p * (p - self.a) * (p - self.b) * (p - self.c))


a, b, c = map(int, input().split())
t = Triangle(a, b, c)
area = t.getArea()

if area == -1:
    print("invalid")
else:
    print(f"{area:.2f}")
```
